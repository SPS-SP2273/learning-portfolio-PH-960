## Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,                            
             color='red', linestyle='dashed',                   # color and linestyle of the plotted straight line
             marker='o', markerfacecolor='yellow',              # marker indicates the location of the plotted points
             ecolor='black',capsize=5,                          # ecolor is the colour of the error bars, capsize denotes the length of the horizontal line at the ends of error bars
             label='Y values with errors')                      # graph label for the legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)                                             # alpha adjusts the opacity/transparency of the inner gridlines
plt.legend(loc='lower right')                                   # loc specifies the location of the legend
plt.show()
```

## Exercise 2 (A simple mathematical plot)

```python
import numpy as np
x = np.linspace(0,51, 300)                                      # x ranges from 0 to 50, with 300 points
y = (np.exp(-1/10*x))*(np.sin(x))                               

plt.plot(x, y, color='blue')
plt.xlabel('x', color = 'darkorange', fontsize = '15')
plt.ylabel('y', color = 'darkorange', fontsize = '15')
plt.ylim([-1,1])                                                # sets y axis limits to -1 and 1
plt.grid(alpha=0.25)
plt.title(r'$y = e^{-x/10}\sin(x)$', fontsize = '20')           # when expressing the exponent's power, use {} instead of ()
plt.show()
```
