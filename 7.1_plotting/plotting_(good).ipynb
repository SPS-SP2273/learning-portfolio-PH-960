### pyplot Interface

```python

from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
plt.plot(x, cos_x, label='$\\cos x$')
plt.plot(x, sin_x, label='$\\sin x$')
plt.legend()
```

### OO Interface

```python

fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,
    figsize=(8,4)
)
# option 1
my_sin = ax[0]
my_cos = ax[1]

#option 2
my_sin,my_cos = ax[0],ax[1]

#option 3
my_sin,my_cos = ax

#option 4
fig, (my_sin, my_cos) = plt.subplots(
    nrows = 2,
    figsize=(8,4)
)

my_sin.plot(x, sin_x, label='$\\sin x$')
my_cos.plot(x, cos_x, label='$\\cos x$')

my_cos.fill_between(x,0,cos_x,alpha = .5)
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,
    figsize=(8,4)
)
# Using friendly variables to access axes
my_sin = ax[0]
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\sin x$')
my_sin.legend()

my_cos.plot(x, cos_x, label='$\\cos x$')
my_cos.legend()

my_sin.fill_between(x,0,sin_x,alpha = .3, color='green')
my_cos.fill_between(x,0,cos_x,alpha = .3, color='green')
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,                        # ncols for columns
    figsize=(8,4)
)
# Using friendly variables to access axes
my_sin = ax[0]
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\sin x$', color = 'r')              # Color choice depends on the purpose of the data
my_sin.legend()

my_cos.plot(x, cos_x, label='$\\cos x$', color = 'g')
my_cos.legend()

my_sin.fill_between(x,0,sin_x,alpha = .3, color='green')
my_cos.fill_between(x,0,cos_x,alpha = .3, color='blue')

plt.suptitle('Trig Functions')

my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')

my_sin.set_ylabel('$\\sin x$')
my_cos.set_ylabel('$\\sin x$')

plt.tight_layout()

#plt.savefig('test2.png', dpi = 300)
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,                        # ncols for columns
    figsize=(8,4),
    sharex='col',
    sharey = 'row'
)
# Using friendly variables to access axes
my_sin = ax[0]
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\sin x$', color = 'r')              # Color choice depends on the purpose of the data
my_sin.legend()
my_cos.set_title('sin')

my_cos.plot(x, cos_x, label='$\\cos x$', color = 'g')
my_cos.legend()
my_cos.set_title('cos')

my_sin.fill_between(x,0,sin_x,alpha = .3, color='green')
my_cos.fill_between(x,0,cos_x,alpha = .3, color='blue')

plt.suptitle('Trig Functions')

my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')

my_sin.set_ylabel('$\\sin x$', rotation = 0)                                    # To make the "sin x" label straight
my_cos.set_ylabel('$\\sin x$', rotation = 0)

plt.tight_layout()

#plt.savefig('test2.png', dpi = 300)
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,                        # ncols for columns
    figsize=(8,4),
    sharex='col',
    sharey = 'row'
)
# Using friendly variables to access axes
my_sin = ax[0]
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\sin x$', color = 'r')              # Color choice depends on the purpose of the data
my_sin.legend()
my_cos.set_title('sin')

my_cos.plot(x, cos_x, label='$\\cos x$', color = 'g')
my_cos.legend()
my_cos.set_title('cos')

my_sin.fill_between(x,0,sin_x,alpha = .3, color='green')
my_cos.fill_between(x,0,cos_x,alpha = .3, color='blue')

for a in [my_sin, my_cos]:                                        # TO create legends for both graphs at once
    a.legend()

plt.suptitle('Trig Functions')

#my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')

my_cos.set_xlim([-np.pi, np.pi])

my_sin.set_ylabel('$\\sin x$', rotation = 0)                                    # To make the "sin x" label straight
my_cos.set_ylabel('$\\sin x$', rotation = 0)

plt.tight_layout()

#plt.savefig('test2.png', dpi = 300)
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2,                        # ncols for columns
    figsize=(8,4),
    sharex='col',
    sharey = 'row'
)
# Using friendly variables to access axes
my_sin = ax[0]
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\sin x$', color = 'r')              # Color choice depends on the purpose of the data
my_sin.legend()
my_cos.set_title('sin')

my_cos.plot(x, cos_x, label='$\\cos x$', color = 'g')
my_cos.legend()
my_cos.set_title('cos')

my_sin.fill_between(x,0,sin_x,alpha = .3, color='green')
my_cos.fill_between(x,0,cos_x,alpha = .3, color='blue')

for a in [my_sin, my_cos]:                                        # TO create legends for both graphs at once
    a.legend()
    a.set_xlim([-np.pi, np.pi])
plt.suptitle('Trig Functions')

#my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')



my_sin.set_ylabel('$\\sin x$', rotation = 0)                                    # To make the "sin x" label straight
my_cos.set_ylabel('$\\sin x$', rotation = 0)

plt.tight_layout()

#plt.savefig('test2.png', dpi = 300)
```

### Using ax

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2, ncols = 3,                       # ncols for columns
    figsize=(8,4)
)

ax[1,1].plot(x,cos_x)

```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2, ncols = 3,                       # ncols for columns
    figsize=(8,4)
)

for row in range(2):
    for col in range(3):
        ax[row,col].plot(x,col*cos_x)
             
ax[0,0].grid(color='r')
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2, ncols = 3,                       # ncols for columns
    figsize=(8,4)
)

for row in range(2):
    for col in range(3):
        ax[row,col].plot(x,col*cos_x, label=f'{row},{col}')

             
for a in ax.flatten():
    a.grid(alpha = .5)
    a.legend(loc='upper center')
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 2, ncols = 3,                       # ncols for columns
    figsize=(8,4),
    sharex='col', sharey='row'
)

for row in range(2):
    for col in range(3):
        ax[row,col].plot(x,col*cos_x, label=f'{row},{col}')

             
for a in ax.flatten():
    a.grid(alpha = .5)
    a.legend(loc='upper center')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(               # "Fig" changes the entire figure, 'ax' changes the axes
    nrows = 5, ncols = 5,                       # ncols for columns
    figsize=(10,10),
    sharex='col', sharey='row'
)

for row in range(5):
    for col in range(5):
        ax[row,col].plot(x,col*cos_x, label=f'{row},{col}')
        ax[row,col].fill_between(x,col*cos_x,alpha=.3)
             
for a in ax.flatten():
    a.grid(alpha = .5)
    a.legend(loc='upper center')

for a in ax[1].flatten():
    a.set_xlabel('x')

plt.tight_layout()
```
