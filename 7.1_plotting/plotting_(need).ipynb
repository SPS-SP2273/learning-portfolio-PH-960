# 1 A simple plot

```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('ggplot')
time = np.arange(10)                                                            # indicates 10 points plotted
y2 = time**2
y3 = time**3

plt.plot(time,y2,label='$y^2$', linestyle='--',color = '#1b9e77')           #Linestyle makes a dotted line, colour code is obtained from colour brewer
plt.plot(time,y3, label = '$y^3$')
#plt.scatter(time,y)
plt.xlabel('time (s)')
plt.ylabel('test (test)')
plt.title('My First Plot')
plt.grid(alpha=0.25)
plt.legend()
plt.savefig('test.png')                                                       # different image files types have different properties
```

```python
time = np.arange(10)
y2 = time**2
sin_t = np.sin(time)

plt.plot(time,y2,label='$y^2$',color = '#1b9e77')           #Linestyle makes a dotted line, colour code is obtained from colour brewer
plt.plot(time,sin_t, label = r'$\sin t$')                                       #to tell python to ignore \s as an escape sequence
#plt.scatter(time,y)
plt.xlabel('time (s)')
plt.ylabel('test (test)')
plt.title('My First Plot')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```

```python
time = np.arange(10)                                
y2 = time**2
sin_t = np.sin(time)

plt.plot(time,y2,label='$y^2$', c='orange')
plt.ylabel('$y^2$' ,rotation=0)  

another_y = plt.twinx() #Create a new y-axis
plt.plot(time,sin_t, label = r'$\sin x$')                                       
#plt.scatter(time,y)
plt.xlabel('time (s)')
plt.ylabel('test (test)')
plt.title('My First Plot')
plt.grid(alpha=0.25)
plt.legend()
plt.show()
```
