## Exercise 3 (Timing Stuff)

```python
squares = [i**2 for i in range(10)]

print(squares)
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)

print(squares)
```

Both codes above give the same results *(as shown by the list of values printed)*

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

Meaning of the output:
When executed, the system will run the entire code a total of 7 times. Within each run, there are a hundred thousand / one million repeats or loops of the code. The corresponding time taken for all the loops is then used to calculate the mean and standard deviation of the time taken for the code to be executed per loop *(which is given as the first two numbers of the output in microseconds and nanoseconds respectively in this case)*

With regards to the two sets of code shown above, the second code runs in a shorter amount of time than the first code according to timeit, however the timeit test for the second code took longer as it ran more loops than the first code (1,000,000 loops compared to 100,000 loops)
