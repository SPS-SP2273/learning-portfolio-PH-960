### Task 1

```python
import os
import glob
import shutil
```

```python
month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

old_name = 'oct-08_singapore.txt'.split(sep = '-')              # separates 'oct' from '08' by specifying to split at the hyphen symbol
for key in month:
    if old_name[0] == key:                                      # to select the corresponding key   
        old_name[0] = month[key]
        break                                                   # to stop the loop once the name has been correctly changed
        
new_name = '-'.join(old_name)                                   # to join '10' to '08' with a hyphen
print(new_name)
  
```

### Task 2

```python
def rename_my_file(old_file_name):
    old_file_name = old_file_name.split(sep = '-')              # separates 'oct' from '08' by specifying to split at the hyphen symbol
    for key in month:
        if old_file_name[0] == key:                                      # to select the corresponding key   
            old_file_name[0] = month[key]
            break                                                   # to stop the loop once the name has been correctly changed
        
    new_file_name = '-'.join(old_file_name)                               # to join '10' to '08' with a hyphen
    return new_file_name

rename_my_file('jul-08_singapore.txt')
```

### Task 3

```python
old_files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for name in old_files:
    new_file = rename_my_file(name)
    print(f'{name} ----> {new_file}')
```

### Task 4

```python
path = r'os-collaboration-exercise-data/*.txt'
for old_directory in glob.glob(os.path.join(path), recursive = True):
    file_directory = os.path.split(old_directory)[0]
    name = os.path.split(old_directory)[-1]
    novel_file_name = rename_my_file(name)
    new_directory = os.path.join(file_directory, novel_file_name)
    try:
        os.replace(old_directory,new_directory)
        print(f'{old_directory} has been renamed to {new_directory}')
    except FileExistsError:
        print(f'{new_directory} already exists')
```
