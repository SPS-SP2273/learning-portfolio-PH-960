## Exercise 1


### Task 1

```python
import os
import glob
import shutil
```

```python
month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

old_name = 'oct-08_singapore.txt'.split(sep = '-')        # First separate 'oct' from '08_singapore.txt'
old_name[0] = month['oct']                                # Redefine old_name[0] as value '10'
new_name1 = '-'.join(old_name)                            # Joins '10' to '08_singapore.txt' with a '-'
new_name2 = new_name1.split(sep = '_')                    # separates 'singapore.txt' from '10-08'  
new_name2 = '-'.join(new_name2)                           # Join '10-08' to 'singapore.txt' via '-'

print(new_name2)
```

### Task 2

```python
def rename_my_file(old_file_name):
    old_file_name = old_file_name.split(sep = '-')              # separates 'oct' from '08' by specifying to split at the hyphen symbol
    for key in month:
        if old_file_name[0] == key:                                 # to select the corresponding key   
            old_file_name[0] = month[key]
            break                                                   # to stop the loop once the name has been correctly changed
        
    new_file_name1 = '-'.join(old_file_name)                        # to join '10' to '08' with a hyphen
    new_file_name2 = new_file_name1.split(sep = '_')                # to separate '10-08' from 'singapore.txt' at the underscore
    new_file_name2 = '-'.join(new_file_name2)                       # to join '10-08' and 'singapore.txt' with '-'
    
    return new_file_name2

rename_my_file('mar-08_singapore.txt')                              # to test the function
```

### Task 3

```python
old_files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for name in old_files:
    new_file = rename_my_file(name)
    print(f'{name} ----> {new_file}')
```

### Task 4

```python
path = r'os-collaboration-exercise-data/*.txt'
for old_directory in glob.glob(path, recursive = True):     
    file_directory = os.path.split(old_directory)[0]                    # to get 'os-collaboration-exercise-data' - (A)
    name = os.path.split(old_directory)[-1]                             # to get file name
    novel_file_name = rename_my_file(name)                              # to rename file name   - (B)
    new_directory = os.path.join(file_directory, novel_file_name)       # joins A and B together
    try:
        os.replace(old_directory,new_directory)                         # to replace old directory with new directory 
        print(f'{old_directory} has been renamed to {new_directory}')
    except FileExistsError:
        print(f'{new_directory} already exists')
```

### Task 5

```python
city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for city in city_list:                                                                  
    path_to_directory = os.path.join('os-collaboration-exercise-data',city)     # Used os.path.join instead of f'os-collaboration-exercise-data/{city}' , so that the code can be used for non-Windows operating systems
    os.mkdir(path_to_directory)                                                 # creates the new city folder
```

### Task 6

```python
glob.glob('os-collaboration-exercise-data/*cairo.*', recursive = True)          
# "*" is placed in front of cairo to search for files that have any initial wording ahead of cairo (e.g. 01-02_cairo.txt). 
# Without it, the function only searches for files named "Cairo"
# "*" is placed after the dot to search for all file extensions 
```

```python
for file in glob.glob('os-collaboration-exercise-data/*cairo.*', recursive = True):     # 'file' refers to the specific text/png/pdf file
    path_to_destination = os.path.join('os-collaboration-exercise-data','Cairo')        # creates new directory for the 'Cairo' folder based on type of operating system (Windows, Mac etc)
    shutil.copy(file, path_to_destination)                                              # copies the specific file into the folder
    print(f'{file} has been copied to {path_to_destination}')
```

### Task 7

```python
city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for city in city_list:
    city_files = f'os-collaboration-exercise-data/*{city.lower()}.*'                       # creates 'os-collaboration-exercise-data/*CITY.*' which selects for all file types with the respective city name. '.lower()' makes the city name lowercase, just as the file name
    for file in glob.glob(city_files, recursive = True):            
        filename = os.path.split(file)[-1]                                                  # to obtain file name to test for file existence later
        file_address = os.path.join('os-collaboration-exercise-data',city,f'{filename}')    # forms the entire file address (used to test file existence later)
        path_to_destination = os.path.join('os-collaboration-exercise-data',city)           # creates folder director for the city 

        if os.path.exists(file_address):                                                    # to test if the specific file already exists in the folder (test is made particularly for Cairo files since the files were copied in task 6)
            print(f'{file_address} already exists. Proceeding to the next file')            
        else:
            shutil.move(file, path_to_destination)                                          # Moves the file into the city folder
            print(f'{file} has been moved to {path_to_destination}!')
```
